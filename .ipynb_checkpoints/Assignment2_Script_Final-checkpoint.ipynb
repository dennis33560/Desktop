```python
#Setting constants for easier visualisations
barcolor = {'#7B68EE','#FF8C00','#00FF00','#DC143C','#8A2BE2','#8B0000','#FF69B4','#FFA07A','#00FF7F','#00FFFF','#191970','#708090'}
piecolor = {'#BA55D3', '#DA70D6', '#DDA0DD'}
histcolor = {'#F08080'}

#Setting up methods for easier visualisation

#Bar plot method
def BarPlot(x,y):
  plt.figure(figsize=(12,6))
  plt.title("Figure " + str(y) + " : Bar Plot of " + x.capitalize() + " Variable", fontsize = 16)
  plt.xlabel(x.capitalize(), fontsize = 16)
  plt.ylabel("Frequency", fontsize = 16)
  plt.style.use('seaborn-whitegrid')
  ax = bank[x].value_counts(normalize = True).plot(kind = "bar", edgecolor = 'black', fontsize = 14, color = barcolor)
  plt.show()

#Pie chart method    
def Pie(x,y):
  plt.figure(figsize=(6,6))
  plt.title("Figure " + str(y) + " : Pie Chart of " + x.capitalize() + " Variable", fontsize = 16)
  plt.xlabel('', fontsize='large')
  ax = bank[x].value_counts(normalize = True).plot(kind = "pie", startangle=90, 
                                                   colors = piecolor, autopct='%1.1f%%', 
                                                   pctdistance=0.7, labeldistance=1.2,
                                                   textprops={'fontsize': 14}, wedgeprops={"edgecolor":"k"})
  plt.show()

#Histogram method
def Hist(x,y):
  plt.figure(figsize=(12,6))
  plt.title("Figure " + str(y) + " : Histogram of " + x.capitalize() + " Variable", fontsize = 16)
  plt.xlabel(x.capitalize())
  plt.ylabel("Frequency")
  plt.style.use('seaborn-whitegrid')
  ax = bank[x].plot(kind = "hist", edgecolor = 'black', color = histcolor)
  plt.show()

#Histogram method for continuous variable with y variable
def hist2(x,y):
  plt.figure(figsize=(12,6))
  data1 = bank.loc[bank['y']=="no", x] 
  data2 = bank.loc[bank['y']=="yes", x] 
  plt.hist(data1, alpha = 0.3, bins = 30, edgecolor = 'black', color = '#00FFFF')
  plt.hist(data2, bins = 30, edgecolor = 'black', color = '#F08080')
  plt.title("Figure " + str(y) + " : Term Deposit by " + x.capitalize(), fontsize = 16)
  plt.xlabel(x.capitalize(), fontsize = 16)
  plt.ylabel("Frequency", fontsize = 16)
  plt.style.use('seaborn-whitegrid')
  plt.legend(bank['y'].unique(), bbox_to_anchor=(1.05, 1), loc=2, fontsize = 16)
```

```python
#Import packages for classification
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn import preprocessing
import math
from sklearn.metrics import accuracy_score
from sklearn import tree
```

```python
#Changing categorical variables to numerical
for column in bank2.columns:
    if bank2[column].dtype == type(object):
        le = preprocessing.LabelEncoder()
        bank2[column] = le.fit_transform(bank2[column])
```

```python
#50% test split
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.5,random_state=0)

#Choosing n_neighbours
print(math.sqrt(len(y_test)))
#We will choose the nearest odd number of the square root of the y_test length for n_neighbours

#Choosing weight
#We will use a distance weight as that is more applicable to the dataset based on the outcome we are predicting

#Choosing the metric
#We will stick with the default minkowski metric

#Choosing the p value
#The default value of 2 will be chosen as a low value is ideal to keep all features relevant
knn(50, 144, 'distance', 'minkowski', 2)
```

```python
#Start of DT
def dt(num):
    y = bank2.y
    x = bank2.drop(['emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed', 'y', 'month', 'day_of_week'], axis=1)
    X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=num,random_state=0)
    clf = DecisionTreeClassifier()
    fit = clf.fit(X_train, y_train)
    y_pre = fit.predict(X_test)
    cm = confusion_matrix(y_test, y_pre)
    print("Confusion Matrix with Test Split of " + str(num) + ":")
    print(cm)
    print
    print("Classification Report with Test Split of " + str(num) + ":")
    print classification_report(y_test,y_pre)
```
